* Can you tell me the difference between a 'group by' and 'having' clauses 
    
    * Group by: used to organize data. It collapses multiple rows into summary rows based on shared values in specific columns

    * having: used to filter the results of those groups. It acts like a WHERE clause, but specifically for aggregated data.

    ex:
    SELECT symbol, SUM(volume)
    FROM Trades
    WHERE trade_date = '2026-04-12' 
    GROUP BY symbol               
    HAVING SUM(volume) > 500000

* Explain the difference between mutable and immutable types, and name them

    Mutable: The object's state or contents can be changed after it is created without changing its identity (memory address).
        Lists, Dictionarys, sets

    Immutable: The object's state cannot be changed once created. If you try to "modify" an immutable object, Python actually creates a brand-new object in memory with a new address.
        ints/floats/booleans, string, tuples


* Explain the difference between a list and a set. In what

    Lists are ordered/allow duplicates. 
    Sets are unordered/unique. Use a Set for membership testing
    (checking if an ID exists) because lookups are O(1) instead of O(n).

## ✅ Question 1: Filter for greater than average

Write a function that takes a list of stock prices and returns a new
list containing only the prices that are higher than the average price of the entire set.

````python
prices = [40, 10, 25, 60, 13, 25, 4, 100, 34, 20]
def filter_prices(prices):
    if not prices: return []
    avg = sum(prices) / len(prices)
    return [p for p in prices if p > avg]

filter_prices(prices=prices)
````

## ✅ Question 2: Remove Duplicates While Preserving Order

You receive a list of event IDs collected from a message queue. Some of the events are duplicates.

```python
events = [3, 5, 3, 7, 5, 9]
```

**Task:**  
Write a function `dedup_events(events)` that returns a new list with duplicates removed **but preserves the original order**.

**Expected Output:**

````python
# Your solution for Question 2
events = [3, 5, 3, 7, 5, 9]

def dedup_events(events):
    new_events = []
    for i in events:
        if i not in new_events:
            new_events.append(i)

    return new_events

ans = dedup_events(events)
print(ans)
````

## ✅ Question 3. Find Most Frequent Value in a Log File

**Problem:**
You're given a list of strings representing event logs. Write a function to find the most frequent event.

**Example Input:**
```python
logs = [
  "ERROR",
  "INFO",
  "DEBUG",
  "ERROR",
  "INFO",
  "ERROR"
]
```

**Expected Output:**
```python
"ERROR"
```

````python
# got a 9 out of 10


logs = [
  "ERROR",
  "INFO",
  "DEBUG",
  "ERROR",
  "INFO",
  "ERROR"
]

ans = {}
for i in logs:
    if i in ans:
        ans[i] += 1
    else:
        ans[i] = 1

ans
max(ans, key=ans.get)
````

## ✅ Question 4: Simulate a Simple ETL Pipeline

**Problem:**
You’re given a CSV file containing user data in the format:  
`user_id, Lastname, Firstname`

Write Python code to:
- Load the data from a list of strings (simulate reading from file).
- Return a list of dictionaries sorted by `user_id`.
- the dictionarys should names as one value, first name followed by last name

**Example Input:**
```python
rows = [
  "2,Thunder,Alice",
  "4,Vance,Bob",
  "3,Wayne,Charlie",
  "1,Star,Tom"
]
```

**Expected Output:**
```python
[
  {'user_id': 1, 'name': 'Tom Star'},
  {'user_id': 2, 'name': 'Alice Thunder'},
  {'user_id': 3, 'name': 'Charlie Wayne'},
  {'user_id': 4, 'name': 'Vance Bob'}
]
```

````python

rows = [
  "2,Thunder,Alice",
  "4,Vance,Bob",
  "3,Wayne,Charlie",
  "1,Star,Tom" 
]

ans = []
for row in rows:
    user_id, Lname, Fname = row.split(',')
    ans.append({
        'user_id': int(user_id),
        'name': Fname + ' ' + Lname
    })

# Sort by signup_date
ans.sort(key=lambda x: x['user_id'])
ans
````
